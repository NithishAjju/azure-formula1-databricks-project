# Formula1 Project Environment Set UP


1) Create External Location "databrickscourseextdls27_formula1"
- We will use "databricks-course-sc" Storage Credential for authentication of the DataBricks for the ADLS GEN2 Storage
2) Create Catalog "formula1"
3) Create Schemas "landing","bronze","silver","gold". Each for one layer of the architecture.
4) Create Volume Files in the landing Schema

##Access Cloud Storage

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrickscourseextdls27_formula1
URL 'abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)
COMMENT "This is a formula1 project container external location";
```

```sql
%fs ls "abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/landing"
```

## Create Catalog "formula1"

```sql
%sql
CREATE CATALOG formula1
MANAGED LOCATION 'abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/'
COMMENT 'Catalog for Formula 1 project data in Databricks';
```

```sql
Show catalogs;
```

## Create landing,bronze,silver,gold schemas in the formula1 catalog

```sql
CREATE SCHEMA IF NOT EXISTS  formula1.landing;
CREATE SCHEMA IF NOT EXISTS  formula1.bronze
    MANAGED LOCATION "abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/bronze ";
CREATE SCHEMA IF NOT EXISTS  formula1.silver
    MANAGED LOCATION "abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/silver ";
CREATE SCHEMA IF NOT EXISTS  formula1.gold
    MANAGED LOCATION "abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/gold ";   
```

```sql
use catalog formula1;
```

```sql
show schemas;
```

## Create Volume Files

```sql
CREATE EXTERNAL VOLUME formula1.landing.files
LOCATION 'abfss://formula1@databrickscourseextdls27.dfs.core.windows.net/landing';
```
