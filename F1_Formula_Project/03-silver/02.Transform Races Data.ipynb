#Transform Races Data
1. Read bronze races table
2. Keep only the columns required for analytics (Drop url column)
3. Standardize column names using snake_case (circuitId → circuit_id, raceName → race_name)
4. Rename columns to make them more meaningful (date → race_date )
5. Remove duplicate records
6. Transform column values of race_name to Title Case
7. Write the transformed data to silver races table

```python
%run ../00-common/01.environment_config
```

```python
bronze_table = f"{catalog_name}.{bronze_schema}.races"
silver_table = f"{catalog_name}.{silver_schema}.races"
```

Step 1-Read bronze races table

```python
races_df = spark.table(bronze_table)
#                            (OR)
# circuits_df = spark.read.option("versionAsOf",0).table(bronze_table)
```

```python
display(races_df)
```

Step 2- Keep only the columns required for analytics (Drop url column)

```python
from pyspark.sql import functions as F
races_selected_df = races_df.select(
                F.col("season"),
                F.col("round"),
                F.col("raceName"),
                F.col("date"),
                F.col("circuitId"),
                F.col("ingestion_timestamp"),
                F.col("source_file")
)
```

```python
display(races_selected_df)
```

Step 3 & 4:
- Standardize column names using snake_case (circuitId → circuit_id, raceName → race_name)
- Rename columns to make them more meaningful (date → race_date )

```python
"""circuits_renamed_df = (

        circuits_selected_df
            .withColumnRenamed("circuitId","circuit_id")
            .withColumnRenamed("raceName","race_name")
            .withColumnRenamed("date","race_date") 
                                                    )

                      OR                             """
races_renamed_df = (races_selected_df
                       .withColumnsRenamed({
                           "circuitId":"circuit_id",
                           "raceName":"race_name",
                           "date":"race_date"}))

display(races_renamed_df)



```

Step 5- Remove duplicate records

```python
"""circuits_distinct_df = circuits_valid_df.distinct()
                        OR  """
races_distinct_df = races_renamed_df.dropDuplicates(["season","round"])
```

Step 6- Transform column values of race_name and locality to Title Case

```python
races_final_df = (races_distinct_df
                            .withColumn("race_name",F.initcap(F.col("race_name")))

)


```

Step 8- Write the transformed data to silver races table

```python
(
    races_final_df
        .write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(silver_table)
)
```
