# Formula1 Project Environment Set UP


1) Create External Location "databrickscourseextdls27_formula1-incr"
- We will use "databricks-course-sc" Storage Credential for authentication of the DataBricks for the ADLS GEN2 Storage
2) Create Catalog "formula1_incr"
3) Create Schemas "landing","bronze","silver","gold". Each for one layer of the architecture.
4) Create Volume Files in the landing Schema

##Access Cloud Storage

```sql
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrickscourseextdls27_formula1_incr
URL 'abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)
COMMENT "This is a formula1-incr project container external location";
```

```sql
%fs ls "abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/landing"
```

## Create Catalog "formula1_incr"

```sql
%sql
show catalogs;
```

```sql
%sql
CREATE CATALOG formula1_incr
MANAGED LOCATION 'abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/landing/'
COMMENT 'Catalog for Formula 1 Incremental load project data in Databricks';
```

## Create landing,bronze,silver,gold schemas in the formula1_incr catalog

```sql
CREATE SCHEMA IF NOT EXISTS  formula1_incr.landing;
CREATE SCHEMA IF NOT EXISTS  formula1_incr.bronze
    MANAGED LOCATION "abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/bronze ";
CREATE SCHEMA IF NOT EXISTS  formula1_incr.silver
    MANAGED LOCATION "abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/silver ";
CREATE SCHEMA IF NOT EXISTS  formula1_incr.gold
    MANAGED LOCATION "abfss://formula1-incr@databrickscourseextdls27.dfs.core.windows.net/gold ";   
```

```sql
use catalog formula1_incr;
```

```sql
show schemas;
```
